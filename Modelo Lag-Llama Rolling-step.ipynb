```python
import sys
import os
sys.path.append(os.getcwd())

import torch
from gluonts.torch.modules.loss import NegativeLogLikelihood
from gluonts.torch.distributions.studentT import StudentTOutput
torch.serialization.add_safe_globals([NegativeLogLikelihood, StudentTOutput])

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from gluonts.evaluation import make_evaluation_predictions, Evaluator
from gluonts.dataset.pandas import PandasDataset
from lag_llama.gluon.estimator import LagLlamaEstimator
```

```python
ruta = r"C:\Users\Lenovo\DATOS_IPC.xlsx"
df = pd.read_excel(ruta, parse_dates=['Fecha'])
df = df.set_index('Fecha').asfreq('MS')
df['Inflacion'] = df['Inflacion'].astype('float32')

print("Index dtype:", type(df.index), "freq:", df.index.freq, "idx start:", df.index[0], "idx end:", df.index[-1])

prediction_length_roll = 1
context_length = 60
num_samples = 300
batch_size = 32
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Usando device:", device)

ckpt_path = "lag-llama.ckpt"
ckpt = torch.load(ckpt_path, map_location=device)
estimator_args = ckpt["hyper_parameters"]["model_kwargs"]


estimator_roll = LagLlamaEstimator(
    ckpt_path=ckpt_path,
    prediction_length=prediction_length_roll,
    context_length=context_length,
    device=torch.device(device),

    # args del checkpoint
    input_size=estimator_args["input_size"],
    n_layer=estimator_args["n_layer"],
    n_embd_per_head=estimator_args["n_embd_per_head"],
    n_head=estimator_args["n_head"],
    scaling=estimator_args["scaling"],
    time_feat=estimator_args["time_feat"],

    nonnegative_pred_samples=True,

    rope_scaling={
        "type": "linear",
        "factor": max(1.0, (context_length + prediction_length_roll) / estimator_args.get("context_length", context_length)),
    },

    batch_size=batch_size,
    num_parallel_samples=num_samples
)

predictor_roll = estimator_roll.create_predictor(
    estimator_roll.create_transformation(),
    estimator_roll.create_lightning_module()
)


months_2024 = pd.date_range(start="2024-01-01", periods=12, freq='MS')
results = []

for month_start in months_2024:
    train_end = (month_start - pd.DateOffset(months=1))
    train_slice = df[:train_end] 

    print("\n>>> Mes objetivo:", month_start.date(), " - train hasta:", train_end.date(),
          "| len(train_slice)=", len(train_slice))
    print("   train_slice.index.dtype:", type(train_slice.index), "freq:", train_slice.index.freq)

    # Si PeriodIndex, convertir a DatetimeIndex
    if isinstance(train_slice.index, pd.PeriodIndex):
        train_slice = train_slice.copy()
        train_slice.index = train_slice.index.to_timestamp()
        print("   Convertido PeriodIndex -> DatetimeIndex")

    train_slice = train_slice[train_slice.index.notna()].sort_index()

    if len(train_slice) < 2:
        print(f"Advertencia: datos insuficientes antes de {month_start.date()} (len={len(train_slice)}) - se salta.")
        continue

    train_ds = PandasDataset(train_slice, freq="M", target="Inflacion")


    forecast_it, ts_it = make_evaluation_predictions(
        dataset=train_ds,   
        predictor=predictor_roll,
        num_samples=num_samples
    )

    forecasts = list(forecast_it)
    if len(forecasts) == 0:
        print(f"No se obtuvo forecast para {month_start.date()}")
        continue

    f = forecasts[0]
    pred_mean = float(f.mean[0])
    p05 = float(f.quantile(0.05)[0])
    p95 = float(f.quantile(0.95)[0])

    real_val = None
    if month_start in df.index:
        real_val = float(df.loc[month_start, 'Inflacion'])

    results.append({
        "Fecha": month_start,
        "Prediccion_Media": pred_mean,
        "Percentil_5": p05,
        "Percentil_95": p95,
        "Real": real_val
    })

    print(f"Predicho {month_start.date()}: mean={pred_mean:.4f}, 5%={p05:.4f}, 95%={p95:.4f}, real={real_val}")

# Resultados

rolling_df = pd.DataFrame(results).set_index('Fecha')
rolling_df.to_excel("rolling_predictions_2024_lag_llama_pandasdataset.xlsx")
print("Guardado:", "rolling_predictions_2024_lag_llama_pandasdataset.xlsx")

plt.figure(figsize=(12,6))
start_plot = pd.Timestamp("2023-01-01")
end_plot = pd.Timestamp("2024-12-01")

# Serie real 2023–2024
plt.plot(df.loc[start_plot:end_plot].index,
         df.loc[start_plot:end_plot,'Inflacion'],
         label="Inflación Real (2023–2024)",
         color="blue", linewidth=2, marker="o")

# Pronóstico rolling
if not rolling_df.empty:
    plt.plot(rolling_df.index, rolling_df['Prediccion_Media'],
             label="Pronóstico rolling 1 mes (Lag-Llama)",
             color="red", linewidth=2, marker="o")

    # IC
    plt.fill_between(rolling_df.index,
                     rolling_df['Percentil_5'],
                     rolling_df['Percentil_95'],
                     color="red", alpha=0.2, label="IC 90%")

    # Valores reales de 2024
    if rolling_df['Real'].notna().any():
        plt.scatter(rolling_df.index[rolling_df['Real'].notna()],
                    rolling_df['Real'].dropna(),
                    color="blue", s=60, zorder=5)

plt.title("Pronóstico rolling 1 mes para 2024 - Lag-Llama", fontsize=14, fontweight="bold")
plt.xlabel("Fecha")
plt.ylabel("Inflación (%)")
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("pronostico_rolling_2024.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Cálculo de métricas

mask = rolling_df['Real'].notna()
y_true = rolling_df.loc[mask, 'Real'].values
y_pred = rolling_df.loc[mask, 'Prediccion_Media'].values

# MAE
mae = np.mean(np.abs(y_true - y_pred))

# RMSE
rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

# MAPE (%)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"MAPE : {mape:.2f}%")
```
